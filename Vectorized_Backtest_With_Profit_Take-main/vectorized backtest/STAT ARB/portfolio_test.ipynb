```python
from vector_backtest import *
from datetime import *
import matplotlib.pyplot as plt
```

```python
t = datetime.now()

p = Portfolio([["JPM", "WFC"],
               ["AMD", "NVDA"], 
               ["MSFT", "AAPL"], 
               ["NEE", "DUK"], 
               ["BX", "BAM"], 
               ["GOOG", "FB"]], date(2021, 1, 1), date(2022, 3, 1), 30, 1000)
p.makePortfolio()

print(datetime.now()-t)
```

```python
plt.plot(p.portfolio, alpha = 0.35)
plt.plot(p.returns[:-1], color='black', label="Total Return")
plt.xticks(rotation='45')
plt.legend(p.portfolio.columns)
plt.rcParams["figure.figsize"] = (12, 6)
plt.show()
```
